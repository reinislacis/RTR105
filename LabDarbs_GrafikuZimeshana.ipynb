# Laboratorijas darbs 2. Grafika zīmēšana.

Dotā funkcija ir cos(sqrt(x))

No numpy tiek importēts linspace priekš masīviem un cos, sqrt.

```python
from numpy import linspace, cos, sqrt
```

Tiek definētas tās pašas funkcijas, kas bija Teilora rindās.

```python
def f1(x):
    return (-1)**0*x**0/(1)

def f2(x):
    return f1(x) *(-1)*x/(1*2)

def f3(x):
    return f2(x)*(-1)*x/(1*2*3*4)

def f4(x):
    return f3(x) *(-1)*x/(1*2*3*4*5*6)
```

Iegūst masīvu priekš skatiļu vizualizācijas.
Definē funkciju y, kas ir reālā cos(sqrt(x)) funkcija.

```python
x= linspace(0, 20, 200)
y= cos(sqrt(x))
```

Tiek piedefinētas jaunās funkcijas. Tās veidoja pirmīt ar Teilora rindām.

```python
a = f1(x)
b = f2(x)
c = f3(x)
d = f4(x)
```

Definē masīvu, ar kuru rādīs grafiku informāciju. 

```python
legend = []
```

Importē matplotlib pyplot funkciju kā plt.

```python
from matplotlib import pyplot as plt
```

Plt.grid() veido režģi. plt.xlabel un plt.ylabel izveido virsrakstus asīm. plt.title veido virsrakstu visam zīmējumam. plt.plot zīmē attiecīgos grafikus. color piešķir katram grafikam krāsu. plt.show parāda gala rezultātu.

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funkcija $cos(sqrt(x))$ un izvirziijumi rindaa')
plt.plot(x, y, color = "#FF0055")
legend.append("$cos(sqrt(x))$")
plt.plot(x,a,color = "#FF55FF")
legend.append("$a$")                    
plt.plot(x,b,color = "#55FF00")
legend.append("$b$")
plt.plot(x,c,color = "#5500FF")
legend.append("$c$")
plt.plot(x,d,color = "#00FF55")
legend.append("$d$")
plt.legend(legend, loc = 2)
plt.show()

plt.show()
```
